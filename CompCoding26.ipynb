#### Q2. Shuffle the Array

```python
def shuffle(nums,n):
    res = []
    for i in range(n):
        res.append(nums[i])
        res.append(nums[i+n])
    return res

nums = [2,5,1,3,4,7]
n = 3
print(shuffle(nums,n))
```

#### Q3. Max Consecutive Ones

```python
def findMaxConsecutiveOnes(nums):
    result = curr = 0
    for num in nums:
        if num == 1:
            curr += 1
        else:
            result = max(curr,result)
            curr = 0

    return max(result,curr)

nums = [1,1,0,1,1,1]
print(findMaxConsecutiveOnes(nums))
```

#### Q1. Set Mismatch

```python

def findErrorNums(nums):
    s = set()
    for i in nums:
        if i in s:
            repeated = i
        else:
            s.add(i)
    n = len(nums)
    expectedSum = (n * (n+1) )//2
    actualSum = sum(nums)
    missing = repeated + (expectedSum-actualSum)
    return [repeated,missing]

nums = [1,2,2,4]
print(findErrorNums(nums))
```

# Misselenious

### 238. Product of Array Except Self

```python
def prodExceptItself(nums):
  n = len(nums)
  res = [1]*n

  prefix = 1
  for i in range(n):
    res[i] = prefix
    prefix *= nums[i]

  suffix = 1
  for i in range(n-1,-1,-1):
    res[i] *= suffix
    suffix *= nums[i]

  return res

nums = [1,2,3,4]
print(prodExceptItself(nums))
```

### 961. N-Repeated Element in Size 2N Array

```python
from collections import Counter

def repeatedNTimes(nums):
    s = set()
    for i in nums:
        if i in s:
            return i
        s.add(i)

nums = [1,2,3,3]
print(repeatedNTimes(nums))
```

### 217. Contains Duplicate

```python
def containsDup(nums):
  return len(nums) != len(set(nums))

nums = [1,1,2,3]
print(containsDup(nums))
```

# Two Pointer

### 283. Move Zeroes

```python
def moveZeroes(nums):
  left = 0
  for right in range(len(nums)):
    if nums[right]!=0:
      nums[right], nums[left] = nums[left], nums[right]
      left +=1

nums = [0,1,0,3,12]
moveZeroes(nums)
print(nums)
```
