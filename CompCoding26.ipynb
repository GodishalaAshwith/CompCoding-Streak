### 961. N-Repeated Element in Size 2N Array

```python
from collections import Counter

def repeatedNTimes(nums):
    s = set()
    for i in nums:
        if i in s:
            return i
        s.add(i)

nums = [1,2,3,3]
print(repeatedNTimes(nums))
```

# Two Pointer

### 283. Move Zeroes

```python
def moveZeroes(nums):
  left = 0
  for right in range(len(nums)):
    if nums[right]!=0:
      nums[right], nums[left] = nums[left], nums[right]
      left +=1

nums = [0,1,0,3,12]
moveZeroes(nums)
print(nums)
```
