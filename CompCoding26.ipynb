### 48. Rotate Image

```python
def rotateImg(matrix):
  n = len(matrix)
  dup = [row[:] for row in matrix]
  for i in range(n):
    for j in range(n-1,-1,-1):
      matrix[i][n-j-1] = dup[j][i]

matrix = [[1,2,3],[4,5,6],[7,8,9]]
print("Before: ",matrix)
rotateImg(matrix)
print("After: ",matrix)
```

#### Q2. Shuffle the Array

```python
def shuffle(nums,n):
    res = []
    for i in range(n):
        res.append(nums[i])
        res.append(nums[i+n])
    return res

nums = [2,5,1,3,4,7]
n = 3
print(shuffle(nums,n))
```

#### Q3. Max Consecutive Ones

```python
def findMaxConsecutiveOnes(nums):
    result = curr = 0
    for num in nums:
        if num == 1:
            curr += 1
        else:
            result = max(curr,result)
            curr = 0

    return max(result,curr)

nums = [1,1,0,1,1,1]
print(findMaxConsecutiveOnes(nums))
```

#### Q1. Set Mismatch

```python

def findErrorNums(nums):
    s = set()
    for i in nums:
        if i in s:
            repeated = i
        else:
            s.add(i)
    n = len(nums)
    expectedSum = (n * (n+1) )//2
    actualSum = sum(nums)
    missing = repeated + (expectedSum-actualSum)
    return [repeated,missing]

nums = [1,2,2,4]
print(findErrorNums(nums))
```

### 9. Palindrome Number

```python
def isPalindrome(x):
  if x < 0 : return False

  x = str(x)
  i = 0; j = len(x)-1

  while i <= j :
    if x[i]!=x[j]: return False
    i+=1
    j-=1

  return True

x = 121
isPalindrome(x)
```

### 206. Reverse Linked List

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def reverseList(self, head: Optional[ListNode]) -> Optional[ListNode]:
        if not head or not head.next:
            return head

        new_head = self.reverseList(head.next)
        head.next.next = head
        head.next = None
        return new_head
```

### 238. Product of Array Except Self

```python
def prodExceptItself(nums):
  n = len(nums)
  res = [1]*n

  prefix = 1
  for i in range(n):
    res[i] = prefix
    prefix *= nums[i]

  suffix = 1
  for i in range(n-1,-1,-1):
    res[i] *= suffix
    suffix *= nums[i]

  return res

nums = [1,2,3,4]
print(prodExceptItself(nums))
```

### 2807. Insert Greatest Common Divisors in Linked List

```python
# Definition for singly-linked list.
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

from math import gcd

def insertGCD(head):
  curr = head
  while curr.next:
    gNode = ListNode(val= gcd(curr.val,curr.next.val), next= curr.next)
    curr.next = gNode
    curr = curr.next.next
  return head

def iterate(head):
  curr = head
  while curr:
    print(curr.val,end="->")
    curr = curr.next

head = ListNode(18,ListNode(6,ListNode(10,ListNode(3))))
iterate(head)
print()
head = insertGCD(head)
iterate(head)
```

### 3110. Score of a String

```python
def scoreString(s):
  res = 0
  for i in range(len(s) - 1):
    res += abs(ord(s[i]) - ord(s[i+1]))
  return res

s = "hello"
scoreString(s)
```

### 1. Two Sum

```python
def twoSum(self, nums: List[int], target: int) -> List[int]:
    temp = {}
    for i in range(len(nums)):
        n = nums[i]
        x = target - n
        if x not in temp:
            temp[n] = i
        else:
            return [temp[x],i]
```

### 704. Binary Search

```python
class Solution:
    def search(self, nums: List[int], target: int) -> int:
        left = 0
        right = len(nums) - 1

        while left <= right:
            mid = (left + right) // 2
            if nums[mid]==target:
                return mid
            elif nums[mid] > target:
                right = mid -1
            else:
                left = mid + 1

        return -1
```

### 283. Move Zeroes

```python
def moveZeroes(nums):
  left = 0
  for right in range(len(nums)):
    if nums[right]!=0:
      nums[right], nums[left] = nums[left], nums[right]
      left +=1

nums = [0,1,0,3,12]
moveZeroes(nums)
print(nums)
```
