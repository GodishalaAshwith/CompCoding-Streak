# Misselenious

### 238. Product of Array Except Self

```python
def prodExceptItself(nums):
  n = len(nums)
  res = [1]*n

  prefix = 1
  for i in range(n):
    res[i] = prefix
    prefix *= nums[i]

  suffix = 1
  for i in range(n-1,-1,-1):
    res[i] *= suffix
    suffix *= nums[i]

  return res

nums = [1,2,3,4]
print(prodExceptItself(nums))
```

### 961. N-Repeated Element in Size 2N Array

```python
from collections import Counter

def repeatedNTimes(nums):
    s = set()
    for i in nums:
        if i in s:
            return i
        s.add(i)

nums = [1,2,3,3]
print(repeatedNTimes(nums))
```

### 217. Contains Duplicate

```python
def containsDup(nums):
  return len(nums) != len(set(nums))

nums = [1,1,2,3]
print(containsDup(nums))
```

# Two Pointer

### 283. Move Zeroes

```python
def moveZeroes(nums):
  left = 0
  for right in range(len(nums)):
    if nums[right]!=0:
      nums[right], nums[left] = nums[left], nums[right]
      left +=1

nums = [0,1,0,3,12]
moveZeroes(nums)
print(nums)
```
